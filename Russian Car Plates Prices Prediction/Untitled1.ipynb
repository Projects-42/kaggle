```python
def convert_dates(df):
    df["date"] = pd.to_datetime(df["date"])
    df["year"] = df["date"].dt.year.astype(str)
    df["month"] = df["date"].dt.month.map(
        {
            1: "january",
            2: "february",
            3: "march",
            4: "april",
            5: "may",
            6: "june",
            7: "july",
            8: "august",
            9: "september",
            10: "october",
            11: "november",
            12: "december"
        }
    )
    df["day_week"] = df["date"].dt.dayofweek.map(
        {
            0: "monday",
            1: "tuesday",
            2: "wednesday",
            3: "thursday",
            4: "friday",
            5: "saturday",
            6: "sunday"
        }
    )
    return df
    
```

```python
def reverse_codes(d):
    dict = {}
    for v, k in REGION_CODES.items():
        for key in k:
            dict[key] = v
    return dict
regions = reverse_codes(REGION_CODES)
```

```python
def extract_government_info(row):
    for ident, info in GOVERNMENT_CODES.items():
        if (row["letters"] == ident[0]) and (ident[1][0] <= int(row["numbers"]) <= ident[1][1]) and (row["region"] == ident[2]):
            return info[0], info[1], info[2], info[3]
        
    return "not governmental", 0, 0, 0
```

```python
def extract_plate_info(df):
    df[["dept_name", "forb_buy", "adv_road", "significance"]] = df[["letters", "numbers", "region"]].apply(extract_government_info, axis=1, result_type="expand")
    df["region"] = df["region"].map(regions)
    
```

```python
def wrangle(path):
    df = pd.read_csv("train.csv")
    convert_dates(df)
    extract_plate(df)
    extract_plate_info(df)
    return df
```

```python
def extract_plate(df):
    df["letters"] = df["plate"].str[0] + df["plate"].str[4: 6]
    df["numbers"] = df["plate"].str[1: 4]
    df["region"] = df["plate"].str[6:]
```

```python
target_enc = ["letters", "numbers", "region"]
ohe = ["year", "month", "day_week"]
```

```python
X = df.drop(["id", "plate", "date", "dept_name", "price"], axis=1)
X
```

```python
y = df["price"]
y
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge
from category_encoders import TargetEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error
```

```python
ct = ColumnTransformer(transformers=
    [
        ("tar_en", TargetEncoder(), target_enc),
        ("ohe", OneHotEncoder(), ohe)
    ],
    remainder='passthrough'
)
X_targeted = ct.fit_transform(X, y=y)
X_targeted = pd.DataFrame(X_targeted)
```

```python
ohed = ct.named_transformers_["ohe"].get_feature_names()
ohed
```

```python
X_targeted.columns = target_enc + list(ohed) + ["forb_buy", "adv_road", "significance"]
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.heatmap(X_targeted.corr(), ax=ax)
```

```python
linear_regression = Pipeline([
    ("ct", ct),
    ("ss", StandardScaler()),
    ("lr", LinearRegression())
])
```

```python
linear_regression.fit(X, y)
```

```python
def smape(estimator, X, y):
    return 100 / len(y) * sum(abs(y - estimator.predict(X)) / (abs(y) + abs(estimator.predict(X))) * 2)
```

```python
smape(linear_regression, X, y)
```

```python
ridge = Pipeline([
    ("ct", ct),
    ("ss", StandardScaler()),
    ("ridge", Ridge(random_state=42))
])
```

```python
params_ridge = {
    "ridge__alpha": np.arange(0.001, 0.01, 0.001)
}
```

```python
ridge_grid = GridSearchCV(ridge, param_grid=params_ridge, scoring=smape, cv=5, verbose=1, error_score="raise", return_train_score=True, n_jobs=-1)
ridge_grid.fit(X, y)
```

```python
smape(ridge_grid, X, y)
```

```python
smape(ridge.predict(X), y)
```

```python
svr = Pipeline([
    ("ct", ct),
    ("ss", StandardScaler()),
    ("svr", SVR())
])
```

```python
params_svr = {
    "svr__degree": [1, 2, 3],
    "svr__C": np.arange(0.1, 10, 0.1),
    "svr__epsilon": np.arange(0.01, 0.2, 0.01)
}
```

```python
svr_grid = RandomizedSearchCV(
    svr, 
    param_distributions=params_svr, 
    scoring=smape, 
    cv=5, 
    verbose=1, 
    error_score="raise", 
    return_train_score=True, 
    n_jobs=-1, 
    random_state=42, 
    n_iter=1000
)
svr_grid.fit(X, y)
```
